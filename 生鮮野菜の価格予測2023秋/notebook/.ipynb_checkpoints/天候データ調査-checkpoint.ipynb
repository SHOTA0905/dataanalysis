 # 天気データ加工

```python
area_pairs = alldata['area'].unique()

yasai_areas = set()

for area_pair in area_pairs:
    areas = area_pair.split('_')
    yasai_areas |= set(areas)

yasai_areas

wea_areas = weather['area'].unique()
wea_areas
```

```python
area_map = {}

update_area_map = {
    '岩手':'盛岡','宮城':'仙台','静岡':'浜松','沖縄':'那覇','神奈川':'横浜','愛知':'名古屋','茨城':'水戸','北海道':'帯広','各地':'全国',
    '兵庫':'神戸','香川':'高松','埼玉':'熊谷','国内':'全国','山梨':'甲府','栃木':'宇都宮','群馬':'前橋','愛媛':'松山'
}

for yasai_area in yasai_areas:
    if yasai_area not in wea_areas and yasai_area not in update_area_map:
        area_map[yasai_area] = '全国' # 外国の天候は全国にしておく
    else:
        area_map[yasai_area] = yasai_area

area_map = {**area_map, **update_area_map}
area_map
```

```python
alldata['area_new'] = alldata['area'].apply(lambda x: '_'.join([area_map[i] for i in x.split('_')]))
alldata
```

# 価格推移（全体）

```python
import numpy as np
import matplotlib.pyplot as plt
kindList = test['kind'].unique()

df = train.copy()

def func(row):
    dt = datetime.datetime(
        int(str(row['date'])[:4]), 
        int(str(row['date'])[4:6]), 
        int(str(row['date'])[6:8]))
    return dt
df['date_2'] = df.apply(func, axis=1)

def func2(row):
    return str(row['date'])[4:6]
df['month'] = df.apply(func2, axis=1)

for kind in kindList:
    df1 = df[df['kind'] == kind].copy()
    df1 = df1[(df1['year'].isin(range(2019,2024)))]

    # グラフ可視化（散布図）
    # https://di-acc2.com/programming/python/18359/#index_id4
    plt.plot_date(df1['date_2'], df1['mode_price'], label='Time Series Graph')

    # 書式設定
    plt.xlabel("Date")             # X軸ラベル
    plt.ylabel("mode_price")           # Y軸ラベル
    plt.title(kind)       # タイトル
    plt.legend(loc="best")         # 凡例
    plt.gcf().autofmt_xdate()      # X軸値を45度回転
    # plt.savefig("date_graph.jpg")  # 画像保存
    plt.show()                     # グラフ表示
    plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
kindList = test['kind'].unique()

df = train.copy()

def func(row):
    dt = datetime.datetime(
        int(str(row['date'])[:4]), 
        int(str(row['date'])[4:6]), 
        int(str(row['date'])[6:8]))
    return dt
df['date_2'] = df.apply(func, axis=1)

def func2(row):
    return str(row['date'])[4:6]
df['month'] = df.apply(func2, axis=1)

for kind in kindList:
    df1 = df[df['kind'] == kind].copy()
    df1 = df1[(df1['year'].isin(range(2000,2024)))]
    df1 = df1[(df1['month'].isin(['05','06','07']))]

    box, ax = plt.subplots()
    ax = sns.boxplot(
        x='year', 
        y='mode_price', 
        data=df1,
        hue = "month"
    )
    plt.rcParams["figure.figsize"] = (16, 6)
    plt.xticks(rotation=90)
    print(kind)
    plt.show()
```
