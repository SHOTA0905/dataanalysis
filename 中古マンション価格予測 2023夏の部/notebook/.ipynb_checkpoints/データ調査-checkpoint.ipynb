# データの加工

```python
# 最寄駅：距離（分）
def normalize_moyori(moyori):
    if moyori == moyori:
        if moyori == '30分?60分':
            moyori = 45
        elif moyori == '1H?1H30':
            moyori = 75
        elif moyori == '1H30?2H':
            moyori = 105
        elif moyori == '2H?':
            moyori = 120
        moyori = int(float(moyori))
    return moyori
alldata['最寄駅：距離（分）'] = alldata['最寄駅：距離（分）'].apply(lambda x: normalize_moyori(x))

#面積（㎡）
import re
def normalize_menseki(area):
    if area == area:
        area = int(re.sub('m\^2未満|㎡以上', '', str(area)))
    return area
alldata['面積（㎡）'] = alldata['面積（㎡）'].apply(lambda x: normalize_menseki(x))
```

```python
#建築年
def convert_wareki_to_seireki(wareki):
    if wareki == wareki:
        if wareki == '戦前':
            wareki = '昭和20年'
        value = wareki[2:-1]
        if value == '元':
            value = 1
        else:
            value = int(value)
        if '昭和' in wareki:
            seireki = 1925+value
        elif '平成' in wareki:
            seireki = 1988+value
        elif '令和' in wareki:
            seireki = 2018+value
    else:
        seireki = wareki
    return seireki
alldata['建築年'] = alldata['建築年'].apply(lambda x: convert_wareki_to_seireki(x))
```

```python
# 取引時点（取引年と取引月の2変数化）　※1月はじまりのため4月はじまりで直すべきかも
# 取引年
def convert_torihikiNen(torihikijiten):
    torihikiYear = torihikijiten[0:4]
    torihikiYear = int(float(torihikiYear))
    return torihikiYear
alldata['取引年'] = alldata['取引時点'].apply(lambda x: convert_torihikiNen(x))
# 取引月
def convert_torihikiTsuki(torihikijiten):
    torihikiTsuki = 1
    if torihikijiten[5:] == '第1四半期':
        torihikiTsuki = 1
    elif torihikijiten[5:] == '第2四半期':
        torihikiTsuki = 4
    elif torihikijiten[5:] == '第3四半期':
        torihikiTsuki = 7
    elif torihikijiten[5:] == '第4四半期':
        torihikiTsuki = 10
    return torihikiTsuki
alldata['取引月'] = alldata['取引時点'].apply(lambda x: convert_torihikiTsuki(x))

# 建築経過年数
'''
def convert_kenchikuYear(kenchikuYear):
    if kenchikuYear < 0:
        return 0
    elif kenchikuYear > 50:
        return 51
    else:
        return kenchikuYear
'''
alldata['建築経過年数'] = alldata['取引年']-alldata['建築年'] 
#alldata['建築経過年数'] = alldata['建築経過年数'].apply(lambda x: convert_kenchikuYear(x))
```

```python
def madori_convert(row):
    is_plusS = 0
    is_plusK = 0
    madori = row['間取り']
    if madori == madori:
        #+S
        if madori.find('＋Ｓ') != -1:
            is_plusS = 1
            madori = madori.replace('＋Ｓ','')
        #+K
        if madori.find('＋Ｋ') != -1:
            is_plusK = 1
            madori = madori.replace('＋Ｋ','')
        
    return(int(is_plusS),int(is_plusK),madori)

df_1[['is_PlusS', 'is_PlusK', '間取り_conv']] = df_1.apply(madori_convert, axis=1, result_type="expand")
```

```python
fm = df_1['間取り_conv'].value_counts()
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"] = [10,4]  
 
plt.bar(
    fm.index, 
    fm.values)
    
plt.show()

#
box, ax = plt.subplots()
ax = sns.boxplot(
    x='間取り_conv', 
    y='取引価格（総額）_log', 
    order=df_1.groupby('間取り_conv')['取引価格（総額）_log'].mean().sort_values(ascending=False).index,
    data=df_1
)
plt.xticks(rotation=90)
plt.show()
    
#
fm = df_1.groupby('間取り_conv')['取引価格（総額）_log'].mean()-df_1['取引価格（総額）_log'].mean()

plt.rcParams["figure.figsize"] = [10,4.0]  
plt.xticks(rotation=90)
plt.bar(
    fm.index, 
    fm.values)
plt.xticks(rotation=90)
plt.show()
```

```python
def display_bar(expclm, dataFrame):
    fm = dataFrame[expclm].value_counts()
    plt.xticks(rotation=90)
    plt.rcParams["figure.figsize"] = [10,4]  

    plt.bar(
        fm.index, 
        fm.values)

    plt.show()
    
def display_boxplot(expclm, dataFrame):
    box, ax = plt.subplots()
    ax = sns.boxplot(
        x=expclm, 
        y='取引価格（総額）_log', 
        order=dataFrame.groupby(expclm)['取引価格（総額）_log'].mean().sort_values(ascending=False).index,
        data=dataFrame
    )
    plt.xticks(rotation=90)
    plt.show()
    
def display_median_1(expclm, dataFrame):
    fm = dataFrame.groupby(expclm)['取引価格（総額）_log'].mean()-dataFrame['取引価格（総額）_log'].mean()

    plt.rcParams["figure.figsize"] = [10,4.0]  
    plt.xticks(rotation=90)
    plt.bar(
        fm.index, 
        fm.values)
    plt.xticks(rotation=90)
    plt.show()
    
def display_median_2(expclm, dataFrame, dataFrame2):
    fm = dataFrame.groupby(expclm)['取引価格（総額）_log'].mean()-dataFrame2['取引価格（総額）_log'].mean()

    plt.rcParams["figure.figsize"] = [10,4.0]  
    plt.xticks(rotation=90)
    plt.bar(
        fm.index, 
        fm.values)
    plt.xticks(rotation=90)
    plt.show()
    
```

```python
chikuCount = trainSuv['地区名'].value_counts()
shikuchosonCount = trainSuv['市区町村名'].value_counts()

def func(row):
    if row['地区名'] == row['地区名']:
        if chikuCount[row['地区名']] >= 50:
            return row['地区名']
        elif shikuchosonCount[row['市区町村名']] >= 100:
            return row['市区町村名']
        else:
            return row['都道府県名']
    else:
        return row['都道府県名']

from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
trainSuv['conv_市区町村名'] = trainSuv.apply(func, axis=1)
trainSuv['class_市区町村名'] = class_le.fit_transform(trainSuv['conv_市区町村名'])
```
